# **Meeting Summarizer**

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
from typing import Dict, List
import pandas as pd
import json
import time
import re
from wordcloud import WordCloud, STOPWORDS
import tempfile
import matplotlib.pyplot as plt
import io
import numpy as np
from PIL import Image, ImageDraw
import os
```

```python
class ImprovedMeetingSummarizer:
    def __init__(self, model_name: str = "google/flan-t5-small", use_fined_tuned: bool = False):
        """Initialize the summarizer with local models."""
        try:
          if use_fined_tuned:
              model_name = "./fine_tuned_new_model"
              if not os.path.exists(model_name):
                  raise ValueError(f"Fine-tuned model not found at {model_name}")
              else:
                  print("Successfully loaded fine-tuned model")

          self.tokenizer = AutoTokenizer.from_pretrained(model_name)
          self.model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

          self.pipeline = pipeline(
              "text2text-generation",
              model=self.model,
              tokenizer=self.tokenizer
          )
        except ValueError as ve:
          print(f"Validation Error: {ve}")
          return None
        except Exception as e:
          print(f"Unexpected Error: {e}")
          return None

    def _clean_text(self, text: str) -> str:
        """Clean and deduplicate text."""
        text = re.sub(r'([A-Z][a-z]+(?:,\s*[A-Z][a-z]+)*(?:\s+and\s+[A-Z][a-z]+)?)\s*,\s*\1+', r'\1', text)
        lines = text.split('\n')
        cleaned_lines = []
        seen_content = set()

        for line in lines:
            line = line.strip()
            line = re.sub(r'\s+', ' ', line)
            line = re.sub(r'^[-•*]\s*|\d+\.\s*', '', line)

            if line and line not in seen_content:
                seen_content.add(line)
                cleaned_lines.append(line)

        return '\n'.join(cleaned_lines)

    def _extract_specific_items(self, text: str, prefix: str) -> List[str]:
        """Extract and clean specific items using a targeted prompt."""
        prompt = f"{prefix}. Extract only concrete, non-redundant items from: {text}"
        response = self.pipeline(
            prompt,
            max_length=200,
            num_return_sequences=1,
            do_sample=False
        )[0]['generated_text']

        items = []
        seen_items = set()

        for line in response.split('\n'):
            line = line.strip()
            line = re.sub(r'^[-•*]\s*|\d+\.\s*', '', line)

            if not line or line.lower() in seen_items:
                continue

            if re.match(r'^([A-Z][a-z]+\s*(?:,\s*[A-Z][a-z]+)*\s*(?:and\s+[A-Z][a-z]+)?\.?\s*)+$', line):
                continue

            seen_items.add(line.lower())
            items.append(line)

        return items

    def generate_summary(self, transcript: str) -> Dict:
        """Generate a structured summary of the meeting."""
        overview_prompt = "Provide a brief, focused overview of the key outcomes from this meeting:"
        overview = self.pipeline(
            overview_prompt + transcript,
            max_length=200,
            num_return_sequences=1,
            do_sample=False
        )[0]['generated_text']

        components = {
            "action_items": self._extract_specific_items(
                transcript,
                "List only concrete action items with their assignees and deadlines"
            ),
            "decisions": self._extract_specific_items(
                transcript,
                "List only final decisions and agreements made in the meeting"
            ),
            "key_points": self._extract_specific_items(
                transcript,
                "List only the main discussion points and outcomes"
            ),
            "follow_ups": self._extract_specific_items(
                transcript,
                "List only specific follow-up tasks and next steps"
            )
        }

        return {
            "overview": self._clean_text(overview),
            "action_items": components["action_items"],
            "decisions": components["decisions"],
            "key_points": components["key_points"],
            "follow_ups": components["follow_ups"]
        }

    def format_output(self, result: Dict, format: str = "markdown") -> str:
        """Format the results in a clean, readable format."""
        if format == "markdown":
            output = ""
            output += "## Overview\n" + f"{result['overview']}\n\n"
            output += "## Action Items\n" + "\n".join(f"- {item}" for item in result['action_items']) + "\n\n"
            output += "## Decisions\n" + "\n".join(f"- {item}" for item in result['decisions']) + "\n\n"
            output += "## Key Points\n" + "\n".join(f"- {item}" for item in result['key_points']) + "\n\n"
            output += "## Follow-ups\n" + "\n".join(f"- {item}" for item in result['follow_ups']) + "\n\n"

            if 'faq_pairs' in result:
                output += "## Frequently Asked Questions\n\n"
                for faq in result['faq_pairs']:
                    output += f"**Q: {faq['question']}**\n\n"
                    output += f"A: {faq['answer']}\n\n"

            return output
        else:
            return json.dumps(result, indent=2)

    def process_meeting(self, transcript: str, include_faq: bool = True, include_wc: bool = True) -> Dict:
        """Process the meeting transcript and return structured information with FAQs."""
        summary = self.generate_summary(transcript)
        result = {
            "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
            **summary
        }

        if include_faq:
            result["faq_pairs"] = self.generate_faq(transcript)

        if include_wc:
            # Generate word cloud for the 'overview' section
            key_points_text = ' '.join(result["key_points"])
            result["word_cloud"] = self.generate_wordcloud(key_points_text)

        return result


    def generate_wordcloud(self, text:str):
        try:
            # Check if text is valid
          if not text.strip():
              raise ValueError("Input text is empty or contains only whitespace.")

          print(f"Text length for word cloud: {len(text)} characters.")

          # Preprocess the text to ensure meaningful content
          if len(text.split()) < 5:  # Arbitrary threshold for minimum words
              raise ValueError("Input text does not have enough words to generate a WordCloud.")

            # Create a temporary directory for the wordcloud image
          wordcloud = WordCloud(
              width=400,
              height=200,
              stopwords=STOPWORDS,
              background_color="white",
              max_words=100  # Optional limit to most frequent words
          ).generate(text)

          # Save the WordCloud as an image in memory
          image_buffer = io.BytesIO()
          wordcloud.to_image().save(image_buffer, format='PNG')
          image_buffer.seek(0)

          # Convert the image buffer to a numpy array for compatibility with Gradio
          image_array = plt.imread(image_buffer)
          return image_array

        except ValueError as ve:
          print(f"Validation Error: {ve}")
          return None
        except Exception as e:
          print(f"Unexpected Error: {e}")
          return None


    def generate_faq(self, transcript: str) -> List[Dict]:
          """Generate FAQ pairs from the meeting transcript with improved reliability.

          Args:
              transcript (str): The meeting transcript text

          Returns:
              List[Dict]: List of FAQ pairs, each containing 'question' and 'answer' keys
          """
          # Break transcript into smaller chunks if it's too long
          max_chunk_length = 1000
          chunks = [transcript[i:i + max_chunk_length]
                  for i in range(0, len(transcript), max_chunk_length)]

          # More specific and structured prompt
          faq_generation_prompt = """
          Based on this meeting transcript, generate 5 specific questions and answers about:
          1. Key decisions made
          2. Important deadlines
          3. Action items and responsibilities
          4. Next steps
          5. Important clarifications

          Format each question to start with 'What', 'Who', 'When', 'How', or 'Why'.

          Meeting transcript: {chunk}
          """

          all_questions = []
          # Process each chunk to gather potential questions
          for chunk in chunks:
              try:
                  response = self.pipeline(
                      faq_generation_prompt.format(chunk=chunk),
                      max_length=512,  # Increased max length
                      num_return_sequences=1,
                      do_sample=True,  # Enable some randomness
                      temperature=0.7,  # Add some temperature for more natural responses
                      top_p=0.9
                  )[0]['generated_text']
                  # Extract questions using regex pattern
                  questions = re.findall(r'(?:^|\n)(?:Q:|Question:)?\s*((?:What|Who|When|How|Why)[^?]+\?)', response)
                  all_questions.extend(questions)
              except Exception as e:
                  print(f"Error processing chunk: {e}")
                  continue

          # Deduplicate and select top 5 most relevant questions
          seen_questions = set()
          unique_questions = []
          for q in all_questions:
              q_normalized = q.lower().strip()
              if q_normalized not in seen_questions and len(unique_questions) < 5:
                  seen_questions.add(q_normalized)
                  unique_questions.append(q.strip())

          faq_pairs = []
          for question in unique_questions:
              try:
                  # More focused answer generation prompt
                  answer_prompt = f"""
                  Based on this meeting transcript, provide a clear and specific answer to this question:
                  Question: {question}

                  Include only factual information mentioned in the transcript. If the answer contains any
                  dates, deadlines, or names, make sure to include them.

                  Transcript: {transcript[:1000]}  # Use first 1000 chars for context
                  """

                  answer = self.pipeline(
                      answer_prompt,
                      max_length=200,
                      num_return_sequences=1,
                      temperature=0.3,  # Lower temperature for more focused answers
                      do_sample=False
                  )[0]['generated_text']

                  # Clean and validate the answer
                  cleaned_answer = self._clean_text(answer)
                  if len(cleaned_answer) > 10:  # Basic validation
                      faq_pairs.append({
                          "question": question,
                          "answer": cleaned_answer
                      })
              except Exception as e:
                  print(f"Error generating answer for question '{question}': {e}")
                  continue

          # If we didn't get any valid FAQs, try one last time with a simpler approach
          if not faq_pairs:
              try:
                  backup_prompt = "Extract 3 important questions and their answers from this meeting: " + transcript[:500]
                  backup_response = self.pipeline(
                      backup_prompt,
                      max_length=300,
                      num_return_sequences=1,
                      do_sample=False
                  )[0]['generated_text']

                  # Split into Q&A pairs
                  pairs = re.split(r'\n(?:Q:|A:)', backup_response)
                  for i in range(0, len(pairs)-1, 2):
                      if i+1 < len(pairs):
                          question = pairs[i].strip()
                          answer = pairs[i+1].strip()
                          if '?' in question and len(answer) > 10:
                              faq_pairs.append({
                                  "question": question,
                                  "answer": answer
                              })
              except Exception as e:
                  print(f"Backup FAQ generation failed: {e}")

          return faq_pairs


```

# **Regular Model**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import gradio as gr
import json
import time
from wordcloud import WordCloud, STOPWORDS
import tempfile
import matplotlib.pyplot as plt

summarizer = ImprovedMeetingSummarizer()

def process_transcript(input_text=None, file=None, include_faq=None, include_wc=None):
    try:
        # Determine the input source
      if input_text:
          transcript = input_text
      elif file:
          # Read the uploaded file content
          with open(file.name, "r", encoding="utf-8") as f:
              transcript = f.read()
      else:
          return "Please provide input text or upload a file."

        # Generate meeting summary
      result = summarizer.process_meeting(transcript, include_faq=include_faq, include_wc=include_wc)
      formatted_output = summarizer.format_output(result, format="markdown")
      wordcloud_image = result.get("word_cloud") if include_wc else None

      return formatted_output, wordcloud_image
    except Exception as e:
        return f"An error occurred while processing the input: {str(e)}"


custom_css = """
.gradio-container {
    background-color: #FFF9C4;  /* Set a Light yellow theme */
    color: #3E2723;            /* Text theme */
    font-family: Arial, sans-serif;
    border-radius: 15px;
    padding: 20px;
}

.gradio-container .title, .gradio-container .description {
    color: #3E2723 !important; /* Ensure titles are visible */
}

.gradio-container input[type='file'] {
    border: 2px solid #3E2723;
    padding: 10px;
    border-radius: 10px;
    color: white;
    background-color: #1E8449; /* Green upload button */
}

"""

# Update the interface to handle generator output
interface = gr.Interface(
    fn=process_transcript,
    inputs=[
        gr.Textbox(lines=10, label="Input Text (Paste Meeting Transcript)"),
        gr.File(label="Upload File (Plain Text Only)"),
        gr.Checkbox(label="Include FAQ", value=True),  # Changed from Include Q&A
        gr.Checkbox(label="Include WordCloud", value=True)
    ],
    outputs=[gr.Markdown(label="Meeting Summary"),gr.Image(label="Word Cloud of Summary")],
    title="Meeting Summarizer",
    description="Enter meeting transcripts as text or upload a file to generate structured summaries, including key points, action items, and FAQ.",
    live=False,
    css=custom_css
)

interface.launch(share=True)
```

# **Fine Tuning**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def preprocess_data(batch):
    # Tokenize input and target texts
    inputs = tokenizer(batch['dialogue'], max_length=512, truncation=True, padding="max_length")
    targets = tokenizer(batch['summary'], max_length=150, truncation=True, padding="max_length")

    # Rename and format
    batch['input_ids'] = inputs['input_ids']
    batch['attention_mask'] = inputs['attention_mask']
    batch['labels'] = targets['input_ids']
    return batch

train_dataset = train_data.map(preprocess_data, batched=True)
valid_dataset = valid_data.map(preprocess_data, batched=True)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=2,
    save_steps=500,
    logging_dir='./logs',
    fp16=True,
    gradient_accumulation_steps=4
)
```

```python
from transformers import Trainer

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset
)

# Start training
trainer.train()
```

```python
# Evaluate the model
results = trainer.evaluate()
print(results)


```

```python
import nltk #Import the nltk library
nltk.download("punkt", quiet=True) # Download the 'punkt' resource for tokenization.
import evaluate #Import the evaluate library
metric = evaluate.load("rouge")
```

```python
def compute_metrics(eval_preds):
   preds, labels = eval_preds

   # decode preds and labels
   labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
   decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
   decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

   # rougeLSum expects newline after each sentence
   decoded_preds = ["\n".join(nltk.sent_tokenize(pred.strip())) for pred in decoded_preds]
   decoded_labels = ["\n".join(nltk.sent_tokenize(label.strip())) for label in decoded_labels]

   result = metric.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)

   return result
```
